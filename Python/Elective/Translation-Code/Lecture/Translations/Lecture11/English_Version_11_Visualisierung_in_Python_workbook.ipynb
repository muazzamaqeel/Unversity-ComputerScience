##### Axis shift

What you would probably call axes are the `spines` in `matplotlib`. There are exactly four of them, namely *left* and *right*, and *top* and *bottom*. To create a "classic" coordinate system, for example, the top and right spines are made invisible and the bottom and left ones are shifted.

- `gca()` provides the axis objects, which also contain the `spines`.

````python
# Classical coordinate system
x = np.linspace(-5, 5, 100)
y = x**2-4
ax = plt.gca()
# make upper and right axis invisible
ax.spines["top"].set_color("none")
ax.spines["right"].set_color("none")
# # move bottom and left axis
ax.spines["bottom"].set_position(("data", 0))
ax.spines["left"].set_position(("data", 0))
#Data means the axis is moved to the data point 0
plt.plot(x, y);
````

### 10.2 Graphical user interfaces with `tkinter`

An interactive program enables communication between humans and computers via a user interface. In the previous lines and cells of code, the user interface had two characteristics:

- The communication is *text-based*, i.e. either the program outputs text or the user writes text via keyboard input
- The communication is *synchronized*. A regulated dialog takes place, the temporal sequence of input and output is precisely defined and predictable. For example, the program waits for an `input()` statement until it has been confirmed with *Enter*.

Modern software systems work differently. Word processing programs and other software have user interfaces (*Graphical User Interface*, *GUI* for short). Such a system has the following characteristics:

- Communication is *multimedia*, i.e. the user only sees the "surface" of the program, which is the interface between man and machine. Inputs are made not only via the keyboard, but also with the mouse. There are a variety of different components, such as input fields, buttons, images, animations.

- Communication is *asynchronous*. Instead of following a predetermined dialog, the user has the freedom to do something at any time. The course of communication is unpredictable. Every user action is called an asynchronous event (*Event*), to which the system reacts in a certain way.

This subsection introduces the module `tkinter`. There are other modules for Python such as `pyGTK`, `wxPython` or `PyQt`, but these will not be discussed in more detail in this lecture due to the high programming effort and resource requirements. The particular advantage of `tkinter` is that it allows you to write functional prototypes for engineering applications with GUI with little programming effort. Because it requires few resources, they also run on a Raspberry Pi, for example.

Developing `tkinter` programs usually involves five steps:

1. import the `tkinter` module (`import tkinter as tk`)

2. create an object for the main GUI window

3. create objects for the controls

4. arrange the controls on the GUI

5. implement an event query

The following table lists a few important controls:

|Control|Description|
|:----|:----|
|Label|Label field and output of text (`string`)|
|Entry|Single-line text field for entering text (`string`)|
|Button| Button for triggering events|
|Frame| Area for placing controls|
|Scale| Slider|
|Canvas|Drawing area for displaying lines, shapes, etc.|
|Radio button|Single selection|
|Check button|Multiple selection|
|Menu| Menu bar|
|Message|Display of multi-line text|
|Scrollbar| Scroll bar for scrollable text, canvas, ...|
|List box|Selection list|

````python
# Text input via Entry
class Login(object):
  def __init__(self):
    self.user = {'Tim':'krone2', 'Mick':'selters$',
                'Laura':'prater00', 'Kai':'pasta$1'} 
    self.fenster = Tk()
    self.ausgabe = Label(self.fenster, width=20, height=2)
    self.name = Entry(self.fenster)                         
    self.passwort = Entry(self.fenster, show='*')           
    self.loginButton = Button(self.fenster,
                      text='login', command=self.login)
    self.name.pack(padx=10, pady=10)                 
    self.passwort.pack(pady=10)
    self.loginButton.pack(pady=10)
    self.ausgabe.pack()
    self.fenster.mainloop()

  def login(self):
    name = self.name.get()                                   
    if name in self.user.keys():
      if self.passwort.get() == self.user[name]:
        self.ausgabe.config(text='Herzlich Willkommen, '+name+'!')
      else: self.ausgabe.config(text='Passwort falsch!')
    else: self.ausgabe.config(text='User unbekannt!')
    self.name.delete(0, END)                   
    self.passwort.delete(0, END)

l = Login()
````

#### 10.2.3 Event processing

In general, an event is something that occurs at an unpredictable time (asynchronously). Typical events are:

- clicking a mouse button
- key combination on the keyboard

There are three problems to be solved for the event principle:

1. Specification of the event. In Python, these are described using *event sequences*.

2. The system must react. Procedures are called using *event handlers*.

3. The event must be linked to handlers. This process is called *binding* in Python.

````python
# Introductory example
class Raster(object):
  def __init__(self):
    liste = [(x,y) for x in range(10) for y in range(10)]
    self.fenster = Tk()
    for (i,j) in liste:
      l=Label(self.fenster, width=2,
                background='white')                       
      l.grid(column=i, row=j)
# Binding the event sequences
      l.bind(sequence='<Button-1>', 
             func=self.linksklick)                  
      l.bind(sequence='<Button-3>', 
             func=self.rechtsklick)
      l.bind(sequence='<Double-Button-1>',
             func=self.doppelklick)
    self.fenster.mainloop()

# Event-Handler
  def linksklick(self, event):
      event.widget.config(bg='black')               

  def rechtsklick(self, event):
      event.widget.config(bg='grey')

  def doppelklick(self, event):
      event.widget.config(bg='white')

r = Raster()
````

````python
# Example: Processing keyboard events
from tkinter import *
class Spielflche(object):
  def __init__(self):
    self.fenster = Tk()
    self.canvas = Canvas(self.fenster, 
                         width='10c', height='5c')    
    self.canvas.pack()                               
    h = PhotoImage(file='python\gemini\erde.gif')
    print(h)
    self.canvas.create_image(0,0,anchor=NW, image=h)  
    g7 = PhotoImage( file='python\gemini\gemini7d.gif')
    self.gemini7 = self.canvas.create_image(
                                   50,50,image=g7)    
    g6 = PhotoImage(file='python\gemini\gemini6.gif')
    self.canvas.create_image(300,30,image=g6)         
    self.fenster.bind('<KeyPress-Down>',self.runter)  
    self.fenster.bind('<KeyPress-Up>', self.rauf)
    self.fenster.bind('<KeyPress-Left>', self.links)
    self.fenster.bind('<KeyPress-Right>', self.rechts)
    self.fenster.mainloop() 
  
# Event-Handler
  def rauf(self, event):                               
    self.canvas.move(self.gemini7, 0, -3)           

  def runter(self, event):
    self.canvas.move(self.gemini7, 0, +3)

  def links(self, event):
    self.canvas.move(self.gemini7, -3, 0)

  def rechts(self, event):
    self.canvas.move(self.gemini7, 3, 0)

s = Spielflche()
````

### 10.3 Integrating `matplotlib` into `tkinter`

Suppose you want to write a Python program that receives measurements from a measuring device and displays them as a diagram in a GUI (*Plotter*). For this, it is useful if you can integrate a `matplotlib` diagram into `tkinter`. 

Below is a small project that connects elements from the worlds of `matplotlib` and `tkinter`. The link is an object of the type `FigureCanvasTkAgg`. It ensures that a Figure object from `matplotlib` can be assigned to a `Tk` object as a Canvas object.

````python
# Example: Recording CO2 measurements
import matplotlib
matplotlib.use('TkAgg')                               
from matplotlib.backends.backend_tkagg import (FigureCanvasTkAgg,
                                               NavigationToolbar2Tk)
from matplotlib.figure import Figure
from tkinter import Tk, Label, Button, Entry, LEFT, BOTH
from time import time

class Plotter:                                       
    def __init__(self):     
        self.f = Figure(figsize=(5,4), dpi=100)       
        self.a = self.f.add_subplot(111)              
        self.a.set_ylim(0, 3000)                      
        self.a.set_xlim(0, 30)
        self.a.grid()                           
        self.t_vector = []                            
        self.c_vector = []                            
        self.start = time()                           
# Widgets
        self.window = Tk()
        self.canvas = FigureCanvasTkAgg(self.f,
                             master=self.window)     
        self.canvas.draw()
        self.canvas.get_tk_widget().pack(fill=BOTH,
                                        expand=True) 
        self.toolbar = NavigationToolbar2Tk(
                          self.canvas, self.window)                                                      
        self.label_c = Label(master=self.window,
                         text=" Kohlendioxid (ppm):")
        self.label_c.pack(side=LEFT)
        self.entry_c = Entry(master=self.window, width=4)
        self.entry_c.pack(side=LEFT)
        self.button = Button(master=self.window, text='Ok',
                             command=self.new_value)
        self.button.pack(side=LEFT)
        self.window.mainloop()

    def new_value(self):                             
        c = float(self.entry_c.get())
        t = (time()-self.start)/60                   
        self.t_vector.append(t)                      
        self.c_vector.append(c)
        self.a.plot(self.t_vector,self.c_vector,"-b.")
        self.canvas.draw()
        self.entry_c.delete(0, END)
Plotter()                                            
````
